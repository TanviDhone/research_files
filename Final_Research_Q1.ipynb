```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_excel('FMCG_Cleaned.xlsx')

# Calculate total net sales for each brand to find the top 5 brands
brand_sales = data.groupby('Brand')['Net Sales'].sum().sort_values(ascending=False).head(5)
top_brands = brand_sales.index.tolist()

# Filter data for the top 5 brands
data_top_brands = data[data['Brand'].isin(top_brands)]

# Aggregate net sales for each brand
brand_aggregated_sales = data_top_brands.groupby('Brand')['Net Sales'].sum()

# Save the results to an Excel file
brand_aggregated_sales_df = brand_aggregated_sales.reset_index()
brand_aggregated_sales_df.to_excel('Top_5_Brands_Sales.xlsx', index=False)

# Plotting aggregated sales for each of the top 5 brands in a bar chart
plt.figure(figsize=(10, 6))
brand_aggregated_sales.plot(kind='bar', color='skyblue')
plt.title('Total Net Sales of Top 5 Brands')
plt.xlabel('Brand')
plt.ylabel('Total Net Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```
